```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
df_schools = pd.read_csv(school_data_to_load)
df_students = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(df_students, df_schools, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#District Summary
#Variables
Total_Schools = df_schools["school_name"].count()
Total_Students = df_students["student_name"].count()
Total_budget = df_schools["budget"].sum()
Average_Math_Score = df_students["math_score"].mean()
Average_Reading_Score = df_students["reading_score"].mean()
Passing_Rate_Math = (df_students[df_students["math_score"] >= 70]["math_score"].count()/df_students["math_score"].count())*100
Passing_Rate_Reading = (df_students[df_students["reading_score"] >= 70]["reading_score"].count()/df_students["reading_score"].count())*100
Overall_Pass_Rate = (Passing_Rate_Math + Passing_Rate_Reading)/2    
```

```python
#create data frame
District_Summary_Raw = pd.DataFrame(
{"Total Schools" : [Total_Schools],
 "Total Students" :[Total_Students],
 "Total budget" : [Total_budget],
 "Average Math Score" : [Average_Math_Score],
 "Average Reading Score" : [Average_Reading_Score],
 "Passing Rate Math" : [Passing_Rate_Math],
 "Passing Rate Reading" : [Passing_Rate_Reading],
 "Overall Pass Rate" : [Overall_Pass_Rate]})
District_Summary = District_Summary_Raw[["Total Schools","Total Students","Total budget","Average Math Score","Average Reading Score", "Passing Rate Math","Passing Rate Reading","Overall Pass Rate"]]
Formatted_District_Summary = District_Summary.style.format({"Total Schools":"{:,}",
                                                            "Total Students":"{:,}",
                                                            "Total budget":"${:,.2f}",
                                                            "Average Math Score":"{:.2f}", 
                                                            "Average Reading Score":"{:.2f}",
                                                            "Passing Rate Math":"{:.2f}%", 
                                                            "Passing Rate Reading":"{:.2f}%", 
                                                            "Overall Passing Rate":"{:.2f}%"})
#Print dataframe
Formatted_District_Summary
```
